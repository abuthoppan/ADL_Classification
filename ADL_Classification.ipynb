### Classification of Acitivities of Dailiy Living from Sensor Data 

```python
import pandas as pd
import numpy as np
from dateutil.parser import parse
import math
from functools import partial
import matplotlib.pyplot as plt
```

```python
# Dataset : http://courses.media.mit.edu/2004fall/mas622j/04.projects/home/
# Loaidng Dataset
raw_subject1_sensors = pd.read_csv("data/subject1/sensors.csv",header=None)
raw_subject1_activities = pd.read_csv("data/subject1/Activities.csv")
raw_subject1_activities_data = pd.read_csv("data/subject1/activities_data.csv",names=range(50),header=None)
raw_subject2_sensors = pd.read_csv("data/subject2/sensors.csv",header=None)
raw_subject2_activities = pd.read_csv("data/subject2/Activities.csv")
raw_subject2_activities_data = pd.read_csv("data/subject2/activities_data.csv",names=range(50),header=None)
```

```python
# Function used as a partial fuction in processData()
def appendDate(date,time):
    return str(date) +" "+ str(time)
```

```python
# Function to process data from the dataset into time series data
def timeSerializeData(activities_data):
    
    temp_series_1 = pd.Series()
    temp_series_2 = pd.Series()
    temp_series_3 = pd.Series()
    temp_series_4 = pd.Series()
    
    sensor_data = pd.DataFrame(columns=['Id','SensorId', 'SensorObject', 'ActivationTime','DeactivationTime'])
    activity_data = pd.DataFrame(columns=['Id','Activity', 'StartTime', 'EndTime'])
    
    temp_sensor_data = []
    
    x = 0
    date_string=""
    for i, row in activities_data.iterrows():
        if i%5 is 1:
            temp_series_1 = pd.Series(row,name="SensorId")
        if i%5 is 2:
            temp_series_2 = pd.Series(row,name="SensorObject")
        if i%5 is 3:
            temp_series_3 = pd.Series(row.apply(partial(appendDate,date_string)),name="ActivationTime")
        if i%5 is 4:
            temp_series_4 = pd.Series(row.apply(partial(appendDate,date_string)),name="DeactivationTime")
        if i%5 is 0:
            activity_data.loc[x] = [x+1,row[0],row[1]+" "+row[2],row[1]+" "+row[3]]
            date_string = row[1]
            if temp_series_1.count() > 0 or temp_series_2.count() > 0 or temp_series_3.count() > 0 or temp_series_4.count() > 0:
                temp_df = pd.concat([pd.Series([x for i in range(0, temp_series_2.count())],name="Id",dtype=np.int64),temp_series_1,temp_series_2,temp_series_3,temp_series_4],axis=1)
                temp_sensor_data.append(temp_df)
            x=x+1

    if temp_series_1.count() > 0 or temp_series_2.count() > 0 or temp_series_3.count() > 0 or temp_series_4.count() > 0:
        temp_df = pd.concat([pd.Series([x for i in range(0, temp_series_2.count())],name="Id",dtype=np.int64),temp_series_1,temp_series_2,temp_series_3,temp_series_4],axis=1)
        temp_sensor_data.append(temp_df)

    sensor_data = pd.concat(temp_sensor_data, ignore_index=True,sort=False)
    sensor_data.dropna(how='any', inplace=True)
    sensor_data = sensor_data.reset_index(drop=True)
    
    activity_timeseries_data = pd.DataFrame(columns=['TimeStamp','Id','Activity', 'Status'])
    x = 0
    for i, row in activity_data.iterrows():
        activity_timeseries_data.loc[x] = [row[2],row[0],row[1],1]
        activity_timeseries_data.loc[x+1] = [row[3],row[0],row[1],0]
        x = x+2

    activity_timeseries_data['TimeStamp'] = pd.to_datetime(activity_timeseries_data['TimeStamp'])
    activity_timeseries_data.set_index('TimeStamp', inplace=True)
    activity_timeseries_data = activity_timeseries_data.sort_index()
    
    
    sensors_timeseries_data = pd.DataFrame(columns=['TimeStamp','ActivityId','SensorId','Sensor', 'Status'])
    x = 0
    for i, row in sensor_data.iterrows():
        sensors_timeseries_data.loc[x] = [row[3],int(row[0]),row[1],row[2],1]
        sensors_timeseries_data.loc[x+1] = [row[4],row[0],row[1],row[2],0]
        x = x+2

    sensors_timeseries_data['TimeStamp'] = pd.to_datetime(sensors_timeseries_data['TimeStamp'])
    sensors_timeseries_data.set_index('TimeStamp', inplace=True)
    sensors_timeseries_data = sensors_timeseries_data.sort_index()
    
    return activity_timeseries_data,sensors_timeseries_data
```

```python
subject1_activity_timeseries_data,subject1_sensors_timeseries_data = timeSerializeData(raw_subject1_activities_data)
subject2_activity_timeseries_data,subject2_sensors_timeseries_data = timeSerializeData(raw_subject2_activities_data)

print(subject1_activity_timeseries_data)
print(subject1_sensors_timeseries_data)
print(subject2_activity_timeseries_data)
print(subject2_sensors_timeseries_data)
```
